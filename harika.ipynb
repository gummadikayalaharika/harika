```python
import pandas as pd
import sqlite3

orders = pd.read_csv("orders.csv")
print(orders.head())
print(orders.info())

users = pd.read_json("users.json")
print(users.head())
print(users.info())

con = sqlite3.connect("restaurants.db")
con.execute("DROP TABLE IF EXISTS restaurants")
with open("restaurants.sql","r") as f:
    sql_script = f.read()
    con.executescript(sql_script)
    
restaurants = pd.read_sql("SELECT * FROM restaurants",con)
print(restaurants.head())
print(restaurants.info())

orders_users = orders.merge(users,on="user_id",how="left")
final_df = orders_users.merge(restaurants,on="restaurant_id",how="left")

print(final_df.head())
print(final_df.shape)
print(final_df.isnull().sum())

final_df.to_csv("final_food_delivery_dataset.csv", index=False)
```
